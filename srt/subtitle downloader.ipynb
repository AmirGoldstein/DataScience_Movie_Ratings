```python
import subprocess
import pysrt
import os
import glob
import time
import pandas as pd
import numpy as np
import re
```

```python
imdb_data = pd.read_csv("imdbData.csv", index_col=0)
imdb_data
```

# Downloading Adult movies srt files:

```python
df_adult = imdb_data[imdb_data['Classification']=='Adult']
```

## runnig bash command to download .srt files with "subgrab"

```python
print("######## Trying to download " + str(len(df_adult)) + " subtitles for movies. ######## ")
for movieName in df_adult.Title:
    bashcommand = f"subgrab -m {movieName} -s -l EN"
    subprocess.Popen(bashcommand).wait()
    print("Trying to download subtitles for: " + movieName)
    time.sleep(3)
```

### Save all .srt file names in list

```python
subNames=[]
i=0
for filename in glob.glob('*.srt'):
    print (i, filename)
    subNames.append(filename)
    i = i+1
```

```python
print(str(len(subNames)) + " subtitles were downloaded succesfully.")
```

### Saving all subtitles text into .txt files

```python
subsArr=[]
for i in range(len(subNames)):
    subs = pysrt.open(f"{subNames[i]}", encoding='utf-8') #loading srt file
    with open(f"adult/moviesText{i}.txt", "w", encoding='utf-8') as text_file:
        print(subs.text,file=text_file)
    print(i,len(subs.text),"words were written to text file")
```
